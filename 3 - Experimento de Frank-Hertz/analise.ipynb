# Experimento de Franck-Hertz

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from scipy.optimize import curve_fit
```

```python
sns.set_theme()
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['figure.dpi'] = 120
```

Dados

```python
data = {} 
```

```python
for file in os.listdir('dados'):
	T, Vr = file.split('_')[1:3]

	T = int(T)
	Vr = float( int(Vr.split('.')[0]) / 100 )

	if T not in data: data[T] = {}
	if Vr not in data[T]: data[T][Vr] = []

	data[T][Vr] = np.loadtxt(f'dados/{file}').transpose()
```

## Curvas de corrente

```python
data.keys()
```

```python
T = 150
plt.title(f'Curvas de corrente para T={T}C')

for key in data[T]:
	sns.lineplot(x=data[150][key][0], y=data[150][key][1], label=f'$V_R$={key}')

plt.xlabel('Tenso (V)')
plt.ylabel('Corrente (A)')
plt.xlim(0, 25)
plt.ylim(0, 2)

plt.show()
```
